```python
from urllib.request import urlopen
from zipfile import ZipFile
from io import BytesIO
import xml.etree.ElementTree as elemTree
```

```python
ApiKey = "444231d722c45a27a3ee191de03c1e7de1c1c096"
corp_code_url = "https://opendart.fss.or.kr/api/corpCode.xml?crtfc_key=" + ApiKey
req_url_format = "https://opendart.fss.or.kr/api/company.json?crtfc_key={}&corp_code={}"
```

```python
resq = urlopen(corp_code_url)
with ZipFile(BytesIO(resq.read())) as zf: 
    file_list = zf.namelist() 
    while len(file_list) > 0: 
        file_name = file_list.pop() 
        corpCode = zf.open(file_name).read().decode()

tree = elemTree.fromstring(corpCode)
XML_stocklist = tree.findall("list") 
corp_code = [x.findtext("corp_code") for x in XML_stocklist] 
corp_name = [x.findtext("corp_name") for x in XML_stocklist] 
stock_code = [x.findtext("stock_code") for x in XML_stocklist] 
modify_date = [x.findtext("modify_date") for x in XML_stocklist] 
stocklist = {} 
for i in range(len(corp_code)): 
    stocklist[corp_code[i]] = (corp_name[i], stock_code[i], modify_date[i])
```

```python
def dart(CorpCode):
    req_header = {"Accept": "text/html, */*; q=0.01",
              "Accept-Encoding": "gzip, deflate",
              "Accept-Language": "ko-KR,ko;q=0.9",
              "Connection": "keep-alive",
              "Content-Length": "18",
              "Content-Type": "application/x-www-form-urlencoded; charset=UTF-8",
              "Cookie": "WMONID=Q8JbsH0LcjG; JSESSIONID=d4jEblg3qzwBc6ymTrbVJoRfpAyhzrkdLpUOWBjImZmqoF9SJGFqWiNpIFTVe6OA.ZG1fZGFydC9EUlQtRFRXQVMwMV9kYXJ0X21zMQ==",
              "Host": "newdart.fss.or.kr",
              "Origin": "http://newdart.fss.or.kr",
              "Referer": "http://newdart.fss.or.kr/dsae001/main.do",
              "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/91.0.4472.124 Safari/537.36",
              'X-Requested-With': "XMLHttpRequest"}
    req_post = requests.post("http://newdart.fss.or.kr/dsae001/select.ax",
                         {"selectKey": CorpCode},req_header)
    soup = BeautifulSoup(req_post.content,"html.parser")
    Temp = [x.text for x in soup.find_all("td")]   
    return Temp[2],Temp[3],Temp[13]
```

```python
excepted = []
for CorpCode in stocklist.keys():
    try:
        Corp_name,Code,Cate = dart(CorpCode)
        dart_ = dart_.append(pd.Series({"Corp Code":CorpCode,
                           "종목코드" : "Not Searched" if Code == " " else Code,
                           "종목명" : Corp_name,
                           "업종명" : Cate}),
            ignore_index=True)
    except :
        excepted.append(CorpCode)
        print("except", stocklist[CorpCode])
```
