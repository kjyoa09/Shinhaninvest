```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
from time import sleep
import json
import urllib.request as req
```

```python
date = [];KRW_USD = []
```

```python
for i in range(1,431):
    url = "https://finance.naver.com/marketindex/exchangeDailyQuote.nhn?marketindexCd=FX_USDKRW&page={}".format(i)
    res = req.urlopen(url).read().decode("CP949")
    soup = BeautifulSoup(res, "html.parser")
    for tmp in soup.find_all("tr")[2:]:
        date.append(tmp.find("td",class_ = "date").text)
        KRW_USD.append(tmp.find("td",class_ = "num").text)
    sleep(1)
    print(date[-1])
```

```python
KRW_USD = pd.DataFrame({"Date":date,
                        "Index":KRW_USD})
KRW_USD["Cate"] = "KRW_USD"
```
